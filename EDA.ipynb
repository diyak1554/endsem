```python
sns.countplot(x='quality',data=df)
plt.title("Distribution of wine quality")
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```

```python
sns.pairplot(df[['alcohol','sulphates','pH','quality']],hue='quality')
```

```python
correlation=df.corr()
print("Top alcohol correlation with quality: ",correlation['quality'].sort_values(ascending=False))
```
